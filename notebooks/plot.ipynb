# Plots for outputs in Nipype and FSL

<mark>**Execute this notebook outside of the docker container**</mark>

In this notebook, we generate plots/figures for Nipype and FSL comparisons that were used in the manuscript.

```python
import numpy as np
from nltools.data import Brain_Data
import matplotlib.pyplot as plt

import nilearn
from nilearn import surface
from nilearn import plotting

fsaverage = nilearn.datasets.fetch_surf_fsaverage()
plotting.show() 
```

## The first-level GLM

Here we randomly choose the four copes from subject-11 run-1

```python
nipype_cope1 = '/srv/scratch/yc/fsl/nipype_output/level1_results/11/results_dir/_subject_id_11/_level1estimate0/results/cope1.nii.gz'
nipype_cope2 = '/srv/scratch/yc/fsl/nipype_output/level1_results/11/results_dir/_subject_id_11/_level1estimate0/results/cope2.nii.gz'
nipype_cope3 = '/srv/scratch/yc/fsl/nipype_output/level1_results/11/results_dir/_subject_id_11/_level1estimate0/results/cope3.nii.gz'
nipype_cope4 = '/srv/scratch/yc/fsl/nipype_output/level1_results/11/results_dir/_subject_id_11/_level1estimate0/results/cope4.nii.gz'

fsl_cope1 = '/srv/scratch/yc/fsl/fsl_gui/three-level_GLM/sub-11/run1.feat/stats/cope1.nii.gz'
fsl_cope2 = '/srv/scratch/yc/fsl/fsl_gui/three-level_GLM/sub-11/run1.feat/stats/cope2.nii.gz'
fsl_cope3 = '/srv/scratch/yc/fsl/fsl_gui/three-level_GLM/sub-11/run1.feat/stats/cope3.nii.gz'
fsl_cope4 = '/srv/scratch/yc/fsl/fsl_gui/three-level_GLM/sub-11/run1.feat/stats/cope4.nii.gz'
```

```python
plotting.plot_stat_map(nipype_cope1, bg_img=nipype_cope1, title = 'Nipype - COPE1', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope1, bg_img=fsl_cope1, title = 'FSL - COPE1', cmap = 'bwr')

plotting.plot_stat_map(nipype_cope2, bg_img=nipype_cope1, title = 'Nipype - COPE2', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope2, bg_img=fsl_cope1, title = 'FSL - COPE2', cmap = 'bwr')

plotting.plot_stat_map(nipype_cope3, bg_img=nipype_cope1, title = 'Nipype - COPE3', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope3, bg_img=fsl_cope1, title = 'FSL - COPE3', cmap = 'bwr')

plotting.plot_stat_map(nipype_cope4, bg_img=nipype_cope1, title = 'Nipype - COPE4', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope4, bg_img=fsl_cope1, title = 'FSL - COPE4', cmap = 'bwr')
```

```python
plotting.plot_stat_map(nipype_cope1, bg_img=nipype_cope1, title = 'Nipype - COPE1', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope1, bg_img=fsl_cope1, title = 'FSL - COPE1', cmap = 'bwr')

plotting.plot_stat_map(nipype_cope2, bg_img=nipype_cope2, title = 'Nipype - COPE2', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope2, bg_img=fsl_cope2, title = 'FSL - COPE2', cmap = 'bwr')

plotting.plot_stat_map(nipype_cope3, bg_img=nipype_cope3, title = 'Nipype - COPE3', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope3, bg_img=fsl_cope3, title = 'FSL - COPE3', cmap = 'bwr')

plotting.plot_stat_map(nipype_cope4, bg_img=nipype_cope4, title = 'Nipype - COPE4', cmap = 'bwr', colorbar = False)
plotting.plot_stat_map(fsl_cope4, bg_img=fsl_cope4, title = 'FSL - COPE4', cmap = 'bwr')
```

```python
brain_fsl_gui = Brain_Data(fsl_cope1).data
print(brain_fsl_gui.shape)
brain_fsl_nipype = Brain_Data(nipype_cope1).data
print(brain_fsl_nipype.shape)

f, a = plt.subplots(nrows=2, figsize=(20, 10), sharex=True)
a[0].plot(brain_fsl_gui, linestyle='-', color='forestgreen')
a[0].set_ylabel('FSL', fontsize=16)
a[0].set_title('COPE1', fontsize=20)
a[1].plot(brain_fsl_nipype, linestyle='-', color='blueviolet')
a[1].set_ylabel('Nipype', fontsize=16)
# a[0].yaxis.set_ticks([])

plt.xlabel('Voxels', fontsize=16)
```

```python
t_nipype_cope1 = surface.vol_to_surf(nipype_cope1, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_nipype_cope1, hemi='right',
        title="Nipype - COPE1", colorbar=False,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)

t_fsl_cope1 = surface.vol_to_surf(fsl_cope1, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_fsl_cope1, hemi='right',
        title="FSL - COPE1", colorbar=True,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)

t_nipype_cope2 = surface.vol_to_surf(nipype_cope2, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_nipype_cope2, hemi='right',
        title="Nipype - COPE2", colorbar=False,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)

t_fsl_cope2 = surface.vol_to_surf(fsl_cope2, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_fsl_cope2, hemi='right',
        title="FSL - COPE2", colorbar=True,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)

t_nipype_cope3 = surface.vol_to_surf(nipype_cope3, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_nipype_cope3, hemi='right',
        title="Nipype - COPE3", colorbar=False,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)

t_fsl_cope3 = surface.vol_to_surf(fsl_cope3, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_fsl_cope3, hemi='right',
        title="FSL - COPE3", colorbar=True,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)

t_nipype_cope4 = surface.vol_to_surf(nipype_cope4, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_nipype_cope4, hemi='right',
        title="Nipype - COPE4", colorbar=False,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)

t_fsl_cope4 = surface.vol_to_surf(fsl_cope4, fsaverage.pial_right)
plotting.plot_surf_stat_map(
        fsaverage.infl_right, t_fsl_cope4, hemi='right',
        title="FSL - COPE4", colorbar=True,
        cmap = 'bwr', bg_map=fsaverage.sulc_right)
```
